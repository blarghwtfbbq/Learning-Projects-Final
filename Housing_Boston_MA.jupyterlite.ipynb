```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('boston_housing.csv')

# Boxplot for MEDV
plt.figure(figsize=(5, 3))
sns.boxplot(x=df['MEDV'])
plt.title('Boxplot of Median Value of Owner-Occupied Homes')
plt.xlabel('MEDV ($1000s)')
plt.show()
```

```python
# Bar plot for CHAS
plt.figure(figsize=(5, 3))
sns.countplot(x=df['CHAS'])
plt.title('Bar Plot of Charles River Variable')
plt.xlabel('CHAS (1 if tract bounds river; 0 otherwise)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Discretize AGE
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=[0, 35, 70, 100], labels=['<=35', '35-70', '>=70'])

# Boxplot for MEDV vs. AGE
plt.figure(figsize=(5, 3))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=df)
plt.title('Boxplot of MEDV vs. AGE Group')
plt.xlabel('Age Group')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

```python
# Scatter plot for NOX vs. INDUS
plt.figure(figsize=(5, 3))
sns.scatterplot(x='INDUS', y='NOX', data=df)
plt.title('Scatter Plot of NOX vs. INDUS')
plt.xlabel('INDUS (proportion of non-retail business acres per town)')
plt.ylabel('NOX (nitric oxides concentration)')
plt.show()
```

```python
# Histogram for PTRATIO
plt.figure(figsize=(5, 3))
sns.histplot(df['PTRATIO'], bins=10)
plt.title('Histogram of Pupil to Teacher Ratio')
plt.xlabel('PTRATIO')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import ttest_ind

# Load dataset
df = pd.read_csv('boston_housing.csv')

# Separate groups
chas1 = df[df['CHAS'] == 1]['MEDV']
chas0 = df[df['CHAS'] == 0]['MEDV']

# Perform T-test
t_stat, p_val = ttest_ind(chas1, chas0, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
from scipy.stats import f_oneway

# Discretize AGE
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=[0, 35, 70, 100], labels=['<=35', '35-70', '>=70'])

# Perform ANOVA
anova_result = f_oneway(df[df['AGE_GROUP'] == '<=35']['MEDV'],
                        df[df['AGE_GROUP'] == '35-70']['MEDV'],
                        df[df['AGE_GROUP'] == '>=70']['MEDV'])
print(f"F-statistic: {anova_result.statistic}, P-value: {anova_result.pvalue}")
```

```python
from scipy.stats import pearsonr

# Perform Pearson Correlation
corr, p_val = pearsonr(df['NOX'], df['INDUS'])
print(f"Pearson Correlation: {corr}, P-value: {p_val}")
```

```python
import statsmodels.api as sm

# Regression analysis
X = df['DIS']
y = df['MEDV']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
